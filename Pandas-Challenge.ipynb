```python
import pandas as pd

school_data = pd.read_csv(r"C:\Users\lealr\OneDrive\DU Bootcamp\Pandas and Jupyter\Pandas-Challenge\schools_complete.csv")
student_data = pd.read_csv(r"C:\Users\lealr\OneDrive\DU Bootcamp\Pandas and Jupyter\Pandas-Challenge\students_complete.csv")
```

```python
school_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_complete.head()
```

```python
# Counting the total number of unique schools
school_count = school_complete['school_name'].nunique()
print(school_count)
```

```python
# Total number of students
student_count = school_complete['student_name'].count()
print(student_count)
```

```python
# The total budget
total_budget = school_data['budget'].sum()
print("$", total_budget)
```

```python
# Average math score
average_math_score = school_complete['math_score'].mean()
print(average_math_score)

# Average reading score
average_reading_score = school_complete['reading_score'].mean()
print(average_reading_score)
```

```python
# Percentage of students passing math
passing_math = school_complete[(school_complete['math_score'] >= 70)].count()["student_name"]
print("Number of students passing math:", passing_math)

passing_math_percentage = passing_math / float(student_count) * 100
print("Percentage of students passing math:", passing_math_percentage)
```

```python
# Percentage of students passing reading
passing_reading = school_complete[(school_complete['reading_score'] >= 70)].count()["student_name"]
print("Number of students passing reading:", passing_reading)


passing_reading_percentage = passing_reading / float(student_count) * 100
print("Percentage of students passing reading:", passing_reading_percentage)
```

```python
# Percentage of students who passed math and reading
passing_students = school_complete[
    (school_complete['math_score'] >= 70) & (school_complete['reading_score'] >= 70)].count()['student_name']

print("Number of passing students:", passing_students)

total_passing_students = passing_students / float(student_count) * 100
print("Percentage of students passing:", total_passing_students)
```

```python
# creating the dataframe
# The pd.DataFrame is a fuction within Pandas that will allow you to create the new dataframe
# Using the functions that I calculated above I'm able to add it to a new dataframe that I labeled 'district_summary'
district_summary = pd.DataFrame({
    'Total Number of Unique Schools': [school_count],
    'Total Number of Students': [student_count],
    'Total Budget': [total_budget],
    'Average Math Score': [average_math_score],
    'Average Reading Score': [average_reading_score],
    '% Passing Math': [passing_math_percentage],
    '% Passing Reading': [passing_reading_percentage],
    'Overall % Passing': [total_passing_students]
})

#using the .map function I'm able to change the format on some of my data
# using the .applymap I'm able to change the format for multiple columns using one line of code
# Need to use a double square bracket when referencing multiple columns
district_summary["Total Number of Students"] = district_summary["Total Number of Students"].map("{:,}".format)
district_summary['Total Budget'] = district_summary['Total Budget']. map("{:,.2f}".format)
district_summary[['% Passing Math', '% Passing Reading', 'Overall % Passing']] = district_summary[['% Passing Math', '% Passing Reading', 'Overall % Passing']]. applymap("{:,.2f}%".format)

district_summary.head()
```
